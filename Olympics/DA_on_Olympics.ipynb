### Seaborn with heatmap

**Using the olympic dataset, present a summary of the total medals won by partipating countries in the 2008 Olympics.**

## Final Challenge

**In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline in your solution**

```python
us = df[df['NOC'] == 'USA'] # US athlete
us
```

```python
usg = us.groupby(['Edition', 'Athlete','Medal', 'Discipline']).size().unstack('Medal', fill_value=0) #olympic year, athlete and discipline grouping
usg
```

```python
usg['Total_m'] = usg['Bronze'] + usg['Gold'] + usg['Silver']
usg
```

```python
usg.reset_index(inplace=True)
usg
```

```python
list = []    
```

```python
for g_key, g_value in usg.groupby('Edition'):
    top_athlete = g_value.sort_values(by='Total_m', ascending=False)[:1]
    list.append(top_athlete)
```

```python
list
```

```python
at = pd.DataFrame()
at = pd.concat(pd.DataFrame(i) for i in list) # concat top 1 athletes
at[['Edition', 'Athlete', 'Discipline', 'Total_m']]
```
