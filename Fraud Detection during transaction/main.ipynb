```python
# Handling data cleaning
def dataCleaning(df):
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)
    df = pd.get_dummies(df, columns=['type'])
    df.drop(columns=['nameOrig', 'nameDest'], axis=1, inplace=True)
    return df
```

```python
cleaned_df = dataCleaning(df)
```

```python
cleaned_df 
```

```python
# Moving isFraud to end
columns = cleaned_df.columns.to_list()

columns.remove('isFraud')
columns.append('isFraud')
cleaned_df = cleaned_df[columns]
```

```python
cleaned_df
```

```python
# Undersampling

fraudulent = cleaned_df[cleaned_df['isFraud'] == 1]
non_fraudulent = cleaned_df[cleaned_df['isFraud'] == 0]

size = fraudulent.shape[0]
non_fraudulent_sampled = non_fraudulent.sample(n=size, random_state=42, axis=0)

balanced_data = pd.concat([fraudulent, non_fraudulent_sampled])

shuffle_data = balanced_data.sample(frac=1, random_state=42)
```
